### 7. Analyze the size of an array
- fruits 샘플의 개수가 몇 개인지 알고 있지만, 실제로는 몇 개의 샘플이 입력될지 알 수 없음
- 따라서 2차원 배열을 1차원 배열로 변환해 샘플의 픽셀 평균값을 계산한다
- Using **`.reshape()`**
- Using **`.mean()`**

```python
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)

print(apple.shape)
print(pineapple.shape)
print(banana.shape)
```

```python
#means of apple
print(apple.mean(axis=1))
```

```python
#means of pineapple
print(pineapple.mean(axis=1))
```

```python
#means of banana
print(banana.mean(axis=1))
```

### 8. Draw a histogram of the overall samples means
- 히스토그램은 전체 데이터의 값의 발생 빈도를 표현
- Using **`.hist()`**
  - axis x: 구간
  - axis y: 발생빈도 수
  - `alpha=`: 그래프 알파 값 지정하여 투명도 조절
- Using **`.legend()`**: 해당 히스토그램이 어떤 데이터를 가리키는지 범례 생성


```python
plt.hist(np.mean(apple, axis=1), alpha=0.8)
plt.hist(np.mean(pineapple, axis=1), alpha=0.8)
plt.hist(np.mean(banana, axis=1), alpha=0.8)
plt.legend(["apple", "pineapple", "banana"])
plt.show()
```

```python
plt.hist(np.mean(apple, axis=1), alpha=0.5)
plt.hist(np.mean(pineapple, axis=1), alpha=0.5)
plt.hist(np.mean(banana, axis=1), alpha=0.5)
plt.legend(["apple", "pineapple", "banana"])
plt.show()
```

### 9. Draw a bar-graph of the overall pixel means

### 10. To output an average image
- 각 픽셀의 해당하는 평균 이미지의 모든 이미지를 출력
- Using **`.subplots()`**
- Using **`.imshow()`**
- Using **`.show()`**

```python
apple_mean = np.mean(apple, axis=0).reshape(100, 100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100, 100)
banana_mean = np.mean(banana, axis=0).reshape(100, 100)

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].imshow(apple_mean, cmap="gray_r")
axs[1].imshow(pineapple_mean, cmap="gray_r")
axs[2].imshow(banana_mean, cmap="gray_r")
plt.show()
```

### 11. To output an image close to the means
- Using **`.subplots()`**
  - 서브 그래프를 생성하고, `figsize` 조정을 통해 그래프 크기 설정
- Using **`.axis()`**
  - `.axis("off")`: 좌표 축 제거 가능
  - `.axis("on")`: 좌표 축 표현
- Using **`.abs()`**: 절댓값을 계산하여 반환하는 함수
- Using **`.argsort()`**: 작은 것에서 큰 순서대로 나열한 배열의 인덱스를 반환
- 

```python
abs_diff = np.abs(fruits - apple_mean)
abs_mean = np.mean(abs_diff, axis=(1, 2))
print(abs_mean.shape)
```

```python
apple_index = np.argsort(abs_mean)[:100]
fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[apple_index[i*10 + j]], cmap="gray_r")
    axs[i, j].axis("off") #"on"= 좌표 축 생성
plt.show()
```

```python
#pineapple
abs_diff = np.abs(fruits - pineapple_mean)
abs_mean = np.mean(abs_diff, axis=(1, 2))
print(abs_mean.shape)
```

```python
pineapple_index = np.argsort(abs_mean)[:100]
fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[pineapple_index[i*10 + j]], cmap="gray_r")
    axs[i, j].axis("on")
plt.show()
```

```python
#banana
abs_diff = np.abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff, axis=(1, 2))
print(abs_mean.shape)
```

```python
banana_index = np.argsort(abs_mean)[:100]
fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[banana_index[i*10 + j]], cmap="gray_r")
    axs[i, j].axis("off") #"on"= 좌표 축 생성
plt.show()
```
