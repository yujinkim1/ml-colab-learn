## Gradient Descent
- Steepest descent(가장 가파른 경사 하강)법이라고도 부른다.
  - 함수 값이 낮아지는 방향으로 독립 변수 값을 변형
  - 최종적으로는 최소 함수 값을 갖도록 하는 독립 변수 값을 찾는 방법
- **왜 미분계수가 0인 지점을 찾는게 아닌 함수의 최솟값을 찾는것을 목적으로 하는가?**
  - 주로 실제 분석에서 사용되는 함수들은 닫힌 형태(closed form)가 아닐 수 있음
  - 함수의 형태가 복잡해 미분계수와 해당하는 근을 계산하기 어려운 경우가 많음
- 경사 하강법 종류
  - **Stochastic Gradient Descent(확률적 경사 하강법)**
  - **Mini-Batch Gradient Descent(미니배치 경사 하강법**
  - **Batch Gradient Descent(전체 경사 하강법)**

### 1. Bath Learning
- 오프라인 학습(Offline Learning)
  - **학습한 것을 시스템에 적용하는 것**
  - **적용 후 더 이상의 학습은 이루어지지 않음**
- 가용한 훈련 데이터를 모두 사용해서 모델을 훈련
  - **시간과 자원을 많이 소모하기 때문에 오프라인으로 수행**
- 새로운 데이터 학습이 요구될 경우 전체 데이터를 다시 처음부터 훈련
  - 자원이 제한된 시스템이라면 더더욱 학습을 위해 시간을 소모
  - 데이터의 양이 많아질 수록 문제가 발생할 여지가 있음



### 2. Incremental Learning
- 온라인 학습(Online Learning)
  - 데이터를 **미니배치(Mini-Batch)** 단위로 묶어 순차적으로 훈련을 진행
- 매 순간 학습이 빠르게 진행
  - 따라서 **미니배치(Mini-Batch)**가 도착 하는대로 바로 학습이 가능
- 연속적으로 데이터를 받고 빠른 변화에 적응해야하는 시스템에 적합
  - 학습률(Learning Rate)
    - 학습률이 높을 경우 : **데이터에 빠르게 적응하지만, 예전 데이터를 금방 휘발**
    - 학습률이 낮은 경우 : **관성이 증가해 학습이 느려지고, 새로운 데이터에 둔감해짐**
- 대표적인 점진적 학습으로 **확률적 경사 하강법(Stochastic Gradient Descent)**가 존재


### 3. SGD(Stochastic Gradient Descent)
- 점진적 학습법
- Stochastic: 무작위로, 임의로
- Gradient: 기울기
- Descent: 기울기를 따라 최적의 위치로 하강하는 것
- 학습 구조
  1. 훈련 세트에서 샘플 하나를 무작위로 추출
  2. 가장 가파른 기울기 값을 추적해 하강
  3. 훈련 세트에서 또 다른 샘플 하나를 무작위로 추출
  4. 가장 가파른 기울기를 따라 최적의 위치로 하강
  5. 샘플을 모두 사용할 때까지 반복 수행
- 만약 완전히 하강하는데 도달하지 못했을 경우
  - **훈련 세트에 모든 샘플을 다시 추가한 후 처음부터 훈련 수행**
  - 만족할 수 있는 위치에 도달할 때까지 계속 하강을 진행
  - 훈련 세트의 샘플을 한 번 모두 사용할 경우 **에포크(Epoch) 1 증가**
- 에포크(Epoch): SGD 알고리즘에서 훈련 세트의 샘플을 한 번 씩 모두 사용하는 과정


### 4. MGD(Mini-Batch Gradient Descent)
- 무작위로 1개의 샘플이 아닌 여러 개의 샘플을 선택
- **미니배치(Mini-Batch)** 단위로 묶어 순차적으로 훈련을 진행
- **미니배치(Mini-Batch)** 마다 경사 하강법 수행

### 5. BGD(Batch Gradient Descent)
- SGD, MGD와 달리 전체 샘플을 한 번 기울기를 따라 사용
- 극단적인 하강법
  - 전체 데이터를 사용한다는 점에서 가장 안정적인 하강법
  - 하지만, 데이터의 양이 많아질 수록 시간과 자원이 극도로 소모될 수 있음

### 8. Implementing a classification model using stochastic gradient descent

### 8-1. Prepare to data
- Import `pandas`
- Read URL of CSV file

```python
import pandas as pd
fish = pd.read_csv("https://bit.ly/fish_csv_data")
```

### 8-2. Create new feature
- Species 열을 제외한 나머지 특성으로 새로운 numpy 배열 생성
- Species 열을 사용해 타겟 생성

```python
fish_input = fish[["Weight", "Length", "Diagonal", "Height", "Width"]].to_numpy()
fish_target = fish["Species"].to_numpy()

#Checkout
print(fish_input)
print(fish_target)
```

### 8-3. Create a Train-set and Test-set
- Import sklearn.model_selection
- Using **`train_test_split()`**

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42
)

#Checkout
print(train_input)
print(train_target)
```

### 8-4. Scale standardization
- import `sklearn.preprocessing`
- Using **`StandardScaler`**

```python
from sklearn.preprocessing import StandardScaler
```

```python
#create a StandardScaler object
ss = StandardScaler()
ss.fit(train_input)

#train-set, test-set 표준화
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

#Checkout
print(train_scaled)
print(test_scaled)
```

### 8-5. Create SGD(Stochastic gradient descent) Model
- Import sklearn.linear_model
- Using **`SGDClassifier`**
  - **`loss`**: 손실함수의 종류 지정
  - **`max_iter`**: 최대 에포크(Epoch) 횟수 지정

```python
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss="log", max_iter=10, random_state=42)
```

### 8-6. Start to fit()

```python
sc.fit(train_scaled, train_target)
#ConvergenceWarning()-훈련된 모델이 충분히 수렴하지 않다는 메세지 출력
```

### 8-7. Check the scores

```python
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
#정확도 점수가 낮아 학습 횟수를 늘려줄 필요가 있음
```

### 8-8. Using **`partial_fit()`**
- 객체를 다시 생성하지 않고, 기존 모델에 추가로 훈련이 가능
- 점진적인 학습을 지원하고 있음
- 기준이 없이 무작정 많이 반복하는 것은 좋지 않음

```python
sc.partial_fit(train_scaled, train_target)
```

### 8-9. Check the scores
- partial_fit() method 사용 후 정확도 점수 확인

```python
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
#재학습 후 정확도 점수가 증가
```

### 9. Epoch and Overfiting / Underfiting
- Import `numpy`
- Using `SGDClassifier`
- Using `np.unique()`
- Import `matplotlib.pyplot`
- Draw a epoch and accuracy graph


```python
import numpy as np
sc = SGDClassifier(loss="log", random_state=42)
train_score = []
test_score = []

classes = np.unique(train_target)
```

```python
for _ in range(0, 300): #300번 수행
  sc.partial_fit(train_scaled, train_target, classes=classes)
  train_score.append(sc.score(train_scaled, train_target))
  test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.title("Epoch and accuracy graph")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.plot(train_score)
plt.plot(test_score)
plt.show()
```

### 10. Conclusion
- Epoch and accuracy graph의 결과로 알 수 있는 점
  - 100번 째 에포크 이후로는 훈련 세트와 테스트 세트의 점수 차이가 벌어지고 있음
  - 에포크 초기는 Underfiting 상태였으나, 100번 째 이후로는 Overfiting의 경향을 보임
  - 따라서 위의 모델은 100번 째 에포크가 적절한 횟수로 판단할 수 있음
- SGDClassifier는 일정 에포크 동안 성능 향상이 없으면 자동으로 멈추게 된다
  - **`tol=None`**으로 설정하면 종료 조건을 None으로 부여해 자동으로 멈추지 않게 할 수 있다.

```python
sc = SGDClassifier(loss="log", max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
train_score.append(sc.score(train_scaled, train_target))
test_score.append(sc.score(test_scaled, test_target))

plt.title("100 times Epoch")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.plot(train_score)
plt.plot(test_score)
plt.show()

#Checkout
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### 10-1. Use a hinge

  - `Surport vector machine`이라 부르며, 머신러닝 알고리즘을 위한 손실 함수

```python
sc = SGDClassifier(loss="hinge", max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
train_score.append(sc.score(train_scaled, train_target))
test_score.append(sc.score(test_scaled, test_target))

plt.title("Use a hinge")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.plot(train_score)
plt.plot(test_score)
plt.show()

#Checkout
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
